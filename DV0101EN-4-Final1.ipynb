```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python
df_survey = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv', index_col=0)
print('Data downloaded and read into a dataframe!')
df_survey.head(10)
```

```python
df_surd = df_survey.sort_values(by='Very interested',ascending=False)

df_surd['Very interested'] = round(df_surd['Very interested'] / 2233, 2)
df_surd['Somewhat interested'] = round(df_surd['Somewhat interested'] / 2233, 2)
df_surd['Not interested'] = round(df_surd['Not interested'] / 2233, 2)

df_surd.head()
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
df_surd.plot(kind='bar', figsize=(20, 8), width=0.8, color=['#5cb85c','#5bc0de','#d9534f'], fontsize=14, grid=False)

plt.title('Percentage of Respondents Interests in Data Science Areas', size=16)
plt.axes().set_facecolor('white')

plt.axes().spines['bottom'].set_color('black')
plt.legend(prop={'size': 14})

for index, value in enumerate(df_surd['Very interested']): 
    plt.annotate(value, xy=(index-0.35, value+0.01), size=14)
for index, value in enumerate(df_surd['Somewhat interested']): 
    plt.annotate(value, xy=(index-0.08, value+0.01), size=14)    
for index, value in enumerate(df_surd['Not interested']): 
    plt.annotate(value, xy=(index+0.18, value+0.01), size=14)
plt.show()
```
